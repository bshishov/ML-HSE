```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score as aucroc
from sklearn.metrics import precision_recall_curve as prCurve
#%% 
def sigma_func(x):
    return float(1) / (float(1) + np.exp(-x))

def gradient_descent(X,Y,W,C,k):
    pred = np.sum(X*W,axis=1)  
    pred = np.array([pred, pred]).T; #Scalar Product
    prob = -pred*Y       ; 
    t = 1.0 -1.0/(1.0 + np.exp(prob));
    a    = X*Y; #Scalar XY
    grad = k * sum(a*t)/Y.size;    
    step = grad - C*W;
    Wupd = W + step;
    return Wupd
#%% Part 1, Load Values
data = np.loadtxt('logistic.csv', delimiter=",")
Y = np.zeros((205,1));
Y[:,0]= data[:,0];

print(sum(X*Y) * [2.0, 1.0])
y_true = data[:,0];
X = data[:,1:];
#%%
W =np.array([[0,0]]);
dist = list();
threshold = 1e-5;
k = 0.1;
C = 10
for i in range(1,10000):
    Wupd = gradient_descent (X,Y,W,C,k);
    d = np.linalg.norm(Wupd-W);
    dist.append(d);
    W = Wupd;
    if (d < threshold):
        break;
print("Stop at %d step" % i)
plt.figure(0)
plt.plot(dist);
yPred = X.dot(W.transpose());
yPredProb = sigma_func(yPred);
plt.figure(1)
plt.plot(yPredProb);
a = aucroc(y_true,yPredProb);
print('AUC_ROC %2.3f' %a)
```
